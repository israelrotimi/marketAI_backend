### Data Cleaning and Preprocessing for NBS Excel Files

```python
cleaned_excel_dataframes = {}

for file_name, df in all_excel_dataframes.items():
    print(f"\nProcessing cleaning for: {file_name}")
    temp_df = df.copy()

    # Check if the first column header is a default 'Unnamed' and the first cell of the first row looks like a header
    # This is a heuristic to decide if the header needs to be shifted down
    # Only apply this if the DataFrame has enough rows and the column names are not already meaningful.
    if not temp_df.empty and len(temp_df) > 0:
        if (isinstance(temp_df.columns[0], str) and temp_df.columns[0].startswith('Unnamed:')) or \
           (isinstance(temp_df.iloc[0, 0], str) and temp_df.iloc[0, 0].strip() in ['Items Label', 'Items', 'ItemLabels', 'NORTH CENTRAL'] and temp_df.columns[0] != temp_df.iloc[0,0]): # Add other common first header values if observed
            temp_df.columns = temp_df.iloc[0]
            temp_df = temp_df[1:].reset_index(drop=True)

        # Drop columns where the header name is NaN (resulting from empty cells in the header row)
        temp_df = temp_df.loc[:, temp_df.columns.notna()]
        # Convert all remaining column names to string type and strip whitespace
        temp_df.columns = temp_df.columns.astype(str).str.strip()

        # Ensure column names are unique (e.g., if first row contained duplicate names)
        cols = []
        seen = {}
        for item in temp_df.columns:
            original_name = str(item).strip()
            name = original_name
            count = 0
            while name in seen:
                count += 1
                name = f"{original_name}_{count}"
            cols.append(name)
            seen[name] = True
        temp_df.columns = cols

    # After setting header, identify and drop rows that are clearly not data
    # For example, the row containing 'NORTH CENTRAL' or entirely NaN/empty rows
    # We'll drop rows where the 'Items Label' is NaN or 'NORTH CENTRAL'
    if 'Items Label' in temp_df.columns:
        temp_df = temp_df[temp_df['Items Label'].notna()]
        temp_df = temp_df[temp_df['Items Label'] != 'NORTH CENTRAL']
        temp_df = temp_df.dropna(how='all') # Drop rows where all values are NaN after initial cleaning
    elif 'Items' in temp_df.columns: # Sometimes the column is just 'Items'
        temp_df = temp_df[temp_df['Items'].notna()]
        temp_df = temp_df[temp_df['Items'] != 'NORTH CENTRAL']
        temp_df = temp_df.dropna(how='all')

    # Identify columns that contain price data (e.g., 'Average of Month-Year')
    # These columns usually start with 'Average of'
    price_columns = [col for col in temp_df.columns if col.startswith('Average of')]

    # Identify id_vars by taking all current columns that are not price columns
    # Ensure id_vars are unique and maintain order
    id_vars = [col for col in temp_df.columns if col not in price_columns]
    id_vars = list(dict.fromkeys(id_vars)) # Ensure uniqueness and preserve order

    # Filter id_vars to only include columns that actually exist in the DataFrame after cleaning
    id_vars = [col for col in id_vars if col in temp_df.columns]

    if price_columns and not temp_df.empty:
        # Melt the DataFrame to transform from wide to long format
        melted_df = pd.melt(temp_df, id_vars=id_vars, value_vars=price_columns, var_name='Month_Year_Avg', value_name='Price')

        # Extract month and year from 'Month_Year_Avg' and create a date column
        # The format is typically 'Average of Mon-YY' or 'Average of Month-YYYY'
        melted_df['Date_Str'] = melted_df['Month_Year_Avg'].str.replace('Average of ', '', regex=False)

        # Attempt to convert to datetime, handling different formats. Coerce errors to NaT
        # Try 'Mon-YY' format first, then 'Month-YYYY'
        melted_df['Date'] = pd.to_datetime(melted_df['Date_Str'], format='%b-%y', errors='coerce')
        melted_df['Date'] = melted_df['Date'].fillna(pd.to_datetime(melted_df['Date_Str'], format='%b-%Y', errors='coerce'))

        # Drop rows where Date could not be parsed
        melted_df = melted_df.dropna(subset=['Date'])

        # Convert Price column to numeric, coercing errors to NaN
        melted_df['Price'] = pd.to_numeric(melted_df['Price'], errors='coerce')
        melted_df = melted_df.dropna(subset=['Price']) # Drop rows where price is NaN after conversion

        # Drop the intermediate 'Month_Year_Avg' and 'Date_Str' columns
        melted_df = melted_df.drop(columns=['Month_Year_Avg', 'Date_Str'], errors='ignore')

        cleaned_excel_dataframes[file_name] = melted_df
        print(f"Cleaned '{file_name}'. First 5 rows:")
        display(melted_df.head())
        print(f"Info for cleaned '{file_name}':")
        melted_df.info()
    elif temp_df.empty:
        print(f"'{file_name}' is empty after initial header processing. Skipping.")
    else:
        print(f"No 'Average of' columns found in '{file_name}'. Skipping melting.")

print("\n\nAll Excel files processed for initial cleaning. Cleaned DataFrames are in 'cleaned_excel_dataframes' dictionary.")
```

### Continued Data Cleaning and Preprocessing for Consolidated Excel Data

I will now standardize the item names, convert relevant columns to numeric types, and extract details from 'Highest' and 'Lowest' columns in the `all_cleaned_food_prices_df`.

```python
# Consolidate item identification columns into a single 'Item' column
all_cleaned_food_prices_df['Item'] = all_cleaned_food_prices_df['Items Label'].fillna(
    all_cleaned_food_prices_df['Items']).fillna(
    all_cleaned_food_prices_df['ItemLabel']).fillna(
    all_cleaned_food_prices_df['ITEMS']).fillna(
    all_cleaned_food_prices_df['Items_1']
)

# Convert MoM and YoY to numeric, coercing errors to NaN
all_cleaned_food_prices_df['MoM'] = pd.to_numeric(all_cleaned_food_prices_df['MoM'], errors='coerce')
all_cleaned_food_prices_df['YoY'] = pd.to_numeric(all_cleaned_food_prices_df['YoY'], errors='coerce')

# Function to extract price and location from 'Highest'/'Lowest' columns
def extract_price_location(text):
    if pd.isna(text):
        return None, None
    text = str(text).strip()
    import re
    match = re.search(r'(.+)\s*\(([-+]?\d*\.?\d+)\)', text)
    if match:
        location = match.group(1).strip()
        price = float(match.group(2))
        return price, location
    # Handle cases where only location or only price might be present, or just a value
    try:
        price = float(text)
        return price, None # Assume it's just a price if convertible
    except ValueError:
        return None, text # Assume it's just a location if not convertible to float

# Apply extraction for 'Highest' and 'Lowest'
all_cleaned_food_prices_df[['Highest_Price', 'Highest_Location']] = all_cleaned_food_prices_df['Highest'].apply(lambda x: pd.Series(extract_price_location(x)))
all_cleaned_food_prices_df[['Lowest_Price', 'Lowest_Location']] = all_cleaned_food_prices_df['Lowest'].apply(lambda x: pd.Series(extract_price_location(x)))

# Drop original, unconsolidated, and highly null columns
columns_to_drop = [
    'Items Label', 'Items', 'ItemLabel', 'ITEMS', 'Items_1',
    'Highest', 'Lowest', # Original columns now replaced by extracted ones
    'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Grand Total'
]
# Also consider dropping regional columns if they were not consistently melted
# For now, I will keep them to inspect, but they might need separate processing
regional_cols = ['NORTH CENTRAL', 'NORTH EAST', 'NORTH WEST', 'SOUTH EAST', 'SOUTH SOUTH', 'SOUTH WEST']

# Drop columns only if they exist in the DataFrame
all_cleaned_food_prices_df = all_cleaned_food_prices_df.drop(
    columns=[col for col in columns_to_drop + regional_cols if col in all_cleaned_food_prices_df.columns]
)

# Convert remaining regional columns (if any) to numeric, coercing errors
for col in regional_cols:
    if col in all_cleaned_food_prices_df.columns:
        all_cleaned_food_prices_df[col] = pd.to_numeric(all_cleaned_food_prices_df[col], errors='coerce')


# Display info and head of the cleaned DataFrame
print("Cleaned and consolidated food prices DataFrame. First 5 rows:")
display(all_cleaned_food_prices_df.head())

print("\nInfo for the cleaned and consolidated DataFrame:")
all_cleaned_food_prices_df.info()
```

### Checking for Semantic Similarity in 'Item' Names

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

# Get unique item names from both dataframes
unique_items_detailed = list(all_cleaned_detailed_prices_df['Item'].unique())
unique_items_food = list(all_cleaned_food_prices_df['Item'].unique())

print("Potential semantic matches for 'Item' names:")
print("-------------------------------------------")

# Iterate through items in the detailed data and find best matches in the food data
potential_matches = []
for item_detailed in unique_items_detailed:
    # Find the best match(es) from unique_items_food
    # scorer=fuzz.token_sort_ratio handles reordering and small differences better
    matches = process.extract(item_detailed, unique_items_food, scorer=fuzz.token_sort_ratio, limit=3)

    # Filter matches that are above a certain similarity threshold (e.g., 70-80)
    filtered_matches = [match for match in matches if match[1] >= 70] # Adjust threshold as needed

    if filtered_matches:
        for match, score in filtered_matches:
            potential_matches.append((item_detailed, match, score))

# Sort by score in descending order
potential_matches.sort(key=lambda x: x[2], reverse=True)

if potential_matches:
    for detailed_item, food_item, score in potential_matches:
        print(f"Detailed: '{detailed_item}' <-> Aggregated: '{food_item}' (Similarity: {score})")
else:
    print("No strong semantic matches found between 'Item' names with a similarity score of 70 or higher.")
```

```python
import numpy as np

# 1. Prepare CPI Data for Merging

# a. Filter cpi_df to only include rows where the Item column is 'Consumer Prices, General Indices (2015 = 100)'.
cpi_filtered_df = cpi_df[cpi_df['Item'] == 'Consumer Prices, General Indices (2015 = 100)'].copy()

# b. Create a YearMonth column in cpi_df by converting its StartDate column to monthly periods
cpi_filtered_df['YearMonth'] = cpi_filtered_df['StartDate'].dt.to_period('M')

# c. Group the filtered cpi_df by YearMonth and calculate the mean Value for each month.
cpi_monthly_avg = cpi_filtered_df.groupby('YearMonth')['Value'].mean().reset_index()

# d. Rename the Value column to cpi_value in this aggregated CPI DataFrame.
cpi_monthly_avg.rename(columns={'Value': 'cpi_value'}, inplace=True)

# e. Calculate the base_cpi for 2024 by taking the mean of cpi_value for all YearMonth values in 2024.
# Ensure 'YearMonth' is treated as period for comparison
cpi_2024 = cpi_monthly_avg[cpi_monthly_avg['YearMonth'].dt.year == 2024]['cpi_value']
if not cpi_2024.empty:
    base_cpi = cpi_2024.mean()
else:
    # Fallback if no 2024 data, or handle as an error if 2024 data is expected.
    # For robustness, we can use the mean of the entire CPI series as a fallback base if 2024 is truly missing
    print("Warning: No CPI data found for 2024. Using overall mean CPI as base.")
    base_cpi = cpi_monthly_avg['cpi_value'].mean()

print(f"Base CPI (2024 average or overall mean if 2024 missing): {base_cpi:.2f}")
print("CPI Monthly Averages (first 5 rows):")
display(cpi_monthly_avg.head())
```

### Generating `market_differentials.csv`

Market differentials indicate the price differences of a commodity across various markets. I will calculate the average price of each commodity in each market and then determine the difference from a national average for that commodity. This will highlight which markets are consistently more expensive or cheaper for certain items.

Steps:
1.  Calculate the national average price per commodity per month.
2.  Calculate the market-specific average price per commodity per month.
3.  Determine the differential as `market_price - national_average_price`.
4.  Save the results to `market_differentials.csv`.

```python
# 1. Calculate national average price per commodity per month
national_avg_prices = analysis_df.groupby(['Date', 'Item'])['price'].mean().reset_index()
national_avg_prices.rename(columns={'price': 'national_avg_price'}, inplace=True)

# 2. Calculate market-specific average price per commodity per month
market_avg_prices = analysis_df.groupby(['Date', 'Item', 'MARKET_SLUG'])['price'].mean().reset_index()
market_avg_prices.rename(columns={'price': 'market_avg_price'}, inplace=True)

# Merge to calculate differentials
market_differentials = pd.merge(market_avg_prices,
                                national_avg_prices,
                                on=['Date', 'Item'],
                                how='left')

# 3. Determine the differential
market_differentials['price_differential'] = market_differentials['market_avg_price'] - market_differentials['national_avg_price']

# Select relevant columns
market_differentials_final = market_differentials[[
    'Date', 'Item', 'MARKET_SLUG', 'market_avg_price', 'national_avg_price', 'price_differential'
]].copy()

# 4. Save to CSV
output_dir = '/content/gdrive/My Drive/data'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

market_differentials_final.to_csv(os.path.join(output_dir, 'market_differentials.csv'), index=False)

print("Market differentials calculated and saved to 'market_differentials.csv'. First 5 rows:")
display(market_differentials_final.head())

print("\nInfo for `market_differentials_final`:")
market_differentials_final.info()
```

### Generating `seasonal_patterns.csv`

Seasonal patterns in food prices are crucial for forecasting. I will analyze the `analysis_df` to identify recurring monthly price variations for each commodity. This involves calculating the average price for each commodity for each month across all years to capture the typical seasonal fluctuations.

Steps:
1.  Extract month and year from the 'Date' column.
2.  Calculate the average price per commodity per month across all markets and years.
3.  Normalize these average prices to show seasonal indices (e.g., percentage deviation from annual average).
4.  Save the results to `seasonal_patterns.csv`.

```python
# Extract month from Date
analysis_df['month'] = analysis_df['Date'].dt.month
analysis_df['year'] = analysis_df['Date'].dt.year

# 2. Calculate the average price per commodity per month
monthly_avg_prices = analysis_df.groupby(['Item', 'month'])['price'].mean().reset_index()
monthly_avg_prices.rename(columns={'price': 'avg_monthly_price'}, inplace=True)

# 3. Calculate overall average price per commodity to normalize (annual average)
commodity_annual_avg = analysis_df.groupby('Item')['price'].mean().reset_index()
commodity_annual_avg.rename(columns={'price': 'overall_avg_price'}, inplace=True)

# Merge to calculate seasonal index
seasonal_patterns = pd.merge(monthly_avg_prices,
                             commodity_annual_avg,
                             on='Item',
                             how='left')

# Calculate seasonal index (deviation from overall average)
seasonal_patterns['seasonal_index'] = (seasonal_patterns['avg_monthly_price'] / seasonal_patterns['overall_avg_price']) * 100

# Select relevant columns
seasonal_patterns_final = seasonal_patterns[[
    'Item', 'month', 'avg_monthly_price', 'seasonal_index'
]].copy()

# 4. Save to CSV
output_dir = '/content/gdrive/My Drive/data'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

seasonal_patterns_final.to_csv(os.path.join(output_dir, 'seasonal_patterns.csv'), index=False)

print("Seasonal patterns calculated and saved to 'seasonal_patterns.csv'. First 5 rows:")
display(seasonal_patterns_final.head())

print("\nInfo for `seasonal_patterns_final`:")
seasonal_patterns_final.info()
```

### Generating `pipeline.py` Script

I will now generate the `pipeline.py` script. This script will consolidate all the data loading, cleaning, and preprocessing logic developed so far. It will include:

1.  **Loading data**: Functions to load CSV and Excel files.
2.  **Cleaning and preprocessing**: Steps for cleaning both the detailed daily data (`all_cleaned_detailed_prices_df`), aggregated monthly data (`all_cleaned_food_prices_df`), and CPI data.
3.  **Real price calculation**: Logic for merging with CPI and calculating real prices.
4.  **Market differential and seasonal pattern calculation**: Functions to generate `market_differentials.csv` and `seasonal_patterns.csv`.
5.  **Database integration**: Code to save the final cleaned price data to `prices_clean.db`.

```python
%%writefile "/content/gdrive/My Drive/pipeline.py"

import pandas as pd
import os
import glob
import re
import sqlite3

def load_dataframes_from_drive(drive_path):
    """Loads all CSV and Excel files from a specified Google Drive path."""
    all_dataframes = {}
    # Load CSVs
    csv_files = glob.glob(os.path.join(drive_path, '*.csv'))
    for file_path in csv_files:
        try:
            file_name = os.path.basename(file_path)
            all_dataframes[file_name] = pd.read_csv(file_path)
        except Exception as e:
            print(f"Error loading CSV {file_name}: {e}")

    # Load Excel files
    excel_files = glob.glob(os.path.join(drive_path, '*.xlsx')) + glob.glob(os.path.join(drive_path, '*.xls'))
    for file_path in excel_files:
        try:
            file_name = os.path.basename(file_path)
            all_dataframes[file_name] = pd.read_excel(file_path)
        except Exception as e:
            print(f"Error loading Excel {file_name}: {e}")

    # Load CSVs from NGA_2025_HFCP_v01_M_CSV folder
    matching_folders = glob.glob(os.path.join(drive_path, 'NGA_2025_HFCP*'))
    if matching_folders:
        folder_path = matching_folders[0]
        folder_csv_files = glob.glob(os.path.join(folder_path, '*.csv'))
        for file_path in folder_csv_files:
            try:
                file_name = os.path.basename(file_path)
                all_dataframes[file_name] = pd.read_csv(file_path)
            except Exception as e:
                print(f"Error loading CSV from folder {file_name}: {e}")

    return all_dataframes

def clean_excel_dataframes(all_excel_dataframes):
    """Cleans and consolidates Excel dataframes."""
    cleaned_excel_dataframes = {}
    for file_name, df in all_excel_dataframes.items():
        temp_df = df.copy()

        if not temp_df.empty and len(temp_df) > 0:
            if (isinstance(temp_df.columns[0], str) and temp_df.columns[0].startswith('Unnamed:')) or \
               (isinstance(temp_df.iloc[0, 0], str) and temp_df.iloc[0, 0].strip() in ['Items Label', 'Items', 'ItemLabels', 'NORTH CENTRAL'] and temp_df.columns[0] != temp_df.iloc[0,0]):
                temp_df.columns = temp_df.iloc[0]
                temp_df = temp_df[1:].reset_index(drop=True)

            temp_df = temp_df.loc[:, temp_df.columns.notna()]
            temp_df.columns = temp_df.columns.astype(str).str.strip()

            cols = []
            seen = {}
            for item in temp_df.columns:
                original_name = str(item).strip()
                name = original_name
                count = 0
                while name in seen:
                    count += 1
                    name = f"{original_name}_{count}"
                cols.append(name)
                seen[name] = True
            temp_df.columns = cols

        if 'Items Label' in temp_df.columns:
            temp_df = temp_df[temp_df['Items Label'].notna()]
            temp_df = temp_df[temp_df['Items Label'] != 'NORTH CENTRAL']
            temp_df = temp_df.dropna(how='all')
        elif 'Items' in temp_df.columns:
            temp_df = temp_df[temp_df['Items'].notna()]
            temp_df = temp_df[temp_df['Items'] != 'NORTH CENTRAL']
            temp_df = temp_df.dropna(how='all')

        price_columns = [col for col in temp_df.columns if col.startswith('Average of')]
        id_vars = [col for col in temp_df.columns if col not in price_columns]
        id_vars = list(dict.fromkeys(id_vars))
        id_vars = [col for col in id_vars if col in temp_df.columns]

        if price_columns and not temp_df.empty:
            melted_df = pd.melt(temp_df, id_vars=id_vars, value_vars=price_columns, var_name='Month_Year_Avg', value_name='Price')
            melted_df['Date_Str'] = melted_df['Month_Year_Avg'].str.replace('Average of ', '', regex=False)
            melted_df['Date'] = pd.to_datetime(melted_df['Date_Str'], format='%b-%y', errors='coerce')
            melted_df['Date'] = melted_df['Date'].fillna(pd.to_datetime(melted_df['Date_Str'], format='%b-%Y', errors='coerce'))
            melted_df = melted_df.dropna(subset=['Date'])
            melted_df['Price'] = pd.to_numeric(melted_df['Price'], errors='coerce')
            melted_df = melted_df.dropna(subset=['Price'])
            melted_df = melted_df.drop(columns=['Month_Year_Avg', 'Date_Str'], errors='ignore')
            cleaned_excel_dataframes[file_name] = melted_df

    all_cleaned_food_prices_df = pd.concat(cleaned_excel_dataframes.values(), ignore_index=True)

    all_cleaned_food_prices_df['Item'] = all_cleaned_food_prices_df['Items Label'].fillna(
        all_cleaned_food_prices_df['Items']).fillna(
        all_cleaned_food_prices_df['ItemLabel']).fillna(
        all_cleaned_food_prices_df['ITEMS']).fillna(
        all_cleaned_food_prices_df['Items_1']
    )
    all_cleaned_food_prices_df['MoM'] = pd.to_numeric(all_cleaned_food_prices_df['MoM'], errors='coerce')
    all_cleaned_food_prices_df['YoY'] = pd.to_numeric(all_cleaned_food_prices_df['YoY'], errors='coerce')

    def extract_price_location(text):
        if pd.isna(text):
            return None, None
        text = str(text).strip()
        match = re.search(r'(.+)\s*\(([-+]?\d*\.?\d+)\)', text)
        if match:
            location = match.group(1).strip()
            price = float(match.group(2))
            return price, location
        try:
            price = float(text)
            return price, None
        except ValueError:
            return None, text

    all_cleaned_food_prices_df[['Highest_Price', 'Highest_Location']] = all_cleaned_food_prices_df['Highest'].apply(lambda x: pd.Series(extract_price_location(x)))
    all_cleaned_food_prices_df[['Lowest_Price', 'Lowest_Location']] = all_cleaned_food_prices_df['Lowest'].apply(lambda x: pd.Series(extract_price_location(x)))

    columns_to_drop = [
        'Items Label', 'Items', 'ItemLabel', 'ITEMS', 'Items_1',
        'Highest', 'Lowest', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Grand Total'
    ]
    regional_cols = ['NORTH CENTRAL', 'NORTH EAST', 'NORTH WEST', 'SOUTH EAST', 'SOUTH SOUTH', 'SOUTH WEST']
    all_cleaned_food_prices_df = all_cleaned_food_prices_df.drop(
        columns=[col for col in columns_to_drop + regional_cols if col in all_cleaned_food_prices_df.columns]
    )

    all_cleaned_food_prices_df['Highest_Price'] = all_cleaned_food_prices_df['Highest_Price'].fillna(all_cleaned_food_prices_df['Price'])
    all_cleaned_food_prices_df['Lowest_Price'] = all_cleaned_food_prices_df['Lowest_Price'].fillna(all_cleaned_food_prices_df['Price'])

    all_cleaned_food_prices_df['COMMODITY_SLUG'] = all_cleaned_food_prices_df['Item'].str.lower().str.replace(' ', '_')
    all_cleaned_food_prices_df['MARKET_SLUG'] = 'nigeria_national_market'

    return all_cleaned_food_prices_df

def clean_detailed_csv_dataframes(all_dataframes):
    """Cleans and consolidates detailed CSV dataframes from HFCP folder."""
    cleaned_detailed_csv_dataframes = {}
    hfcp_csv_files = [k for k in all_dataframes.keys() if k.startswith('level2_price_') or k == 'Raw_data_spatialrev.csv']

    for file_name in hfcp_csv_files:
        df = all_dataframes[file_name].copy()
        df.columns = df.columns.str.lower().str.replace(' ', '_')

        if 'date' in df.columns:
            df['date'] = pd.to_datetime(df['date'], errors='coerce')
            df = df.dropna(subset=['date'])

        for col in ['price', 'daily_mean', 'daily_sd']:
            if col in df.columns:
                df[col] = pd.to_numeric(df[col], errors='coerce')

        subset_for_dropna = []
        if 'price' in df.columns:
            subset_for_dropna.append('price')
        if 'commodity' in df.columns:
            subset_for_dropna.append('commodity')

        if subset_for_dropna:
            df = df.dropna(subset=subset_for_dropna)

        # Create COMMODITY_SLUG for all_cleaned_detailed_prices_df
        df['COMMODITY_SLUG'] = df['commodity'].str.lower().str.replace(' ', '_')

        # Create MARKET_SLUG for all_cleaned_detailed_prices_df
        def create_market_slug(row):
            components = []
            if pd.notna(row['geopol_zon']):
                components.append(str(row['geopol_zon']))
            if pd.notna(row['statelabel']):
                components.append(str(row['statelabel']))
            if pd.notna(row['lgalabel']):
                components.append(str(row['lgalabel']))
            slug = '_'.join(components).lower()
            slug = re.sub(r'[^a-z0-9_]', '', slug)
            return slug

        df['MARKET_SLUG'] = df.apply(create_market_slug, axis=1)

        cleaned_detailed_csv_dataframes[file_name] = df

    all_cleaned_detailed_prices_df = pd.concat(cleaned_detailed_csv_dataframes.values(), ignore_index=True)
    return all_cleaned_detailed_prices_df

def prepare_cpi_data(all_dataframes):
    """Prepares CPI data for real price calculations."""
    cpi_df = all_dataframes['consumer-price-indices_nga.csv'].copy()
    cpi_df['StartDate'] = pd.to_datetime(cpi_df['StartDate'], errors='coerce')
    cpi_df['Value'] = pd.to_numeric(cpi_df['Value'], errors='coerce')
    cpi_df = cpi_df.dropna(subset=['StartDate', 'Value'])
    cpi_df = cpi_df.sort_values(by='StartDate').reset_index(drop=True)

    cpi_filtered_df = cpi_df[cpi_df['Item'] == 'Consumer Prices, General Indices (2015 = 100)'].copy()
    cpi_filtered_df['YearMonth'] = cpi_filtered_df['StartDate'].dt.to_period('M')
    cpi_monthly_avg = cpi_filtered_df.groupby('YearMonth')['Value'].mean().reset_index()
    cpi_monthly_avg.rename(columns={'Value': 'cpi_value'}, inplace=True)

    cpi_2024 = cpi_monthly_avg[cpi_monthly_avg['YearMonth'].dt.year == 2024]['cpi_value']
    if not cpi_2024.empty:
        base_cpi = cpi_2024.mean()
    else:
        base_cpi = cpi_monthly_avg['cpi_value'].mean()

    return cpi_monthly_avg, base_cpi

def prepare_final_prices_df(all_cleaned_detailed_prices_df, all_cleaned_food_prices_df, cpi_monthly_avg, base_cpi):
    """Prepares and consolidates all price data into a single DataFrame."""
    # Prepare HFCP data
    detailed_prices_with_cpi = all_cleaned_detailed_prices_df.copy()
    detailed_prices_with_cpi['YearMonth'] = detailed_prices_with_cpi['date'].dt.to_period('M')
    detailed_prices_with_cpi = pd.merge(
        detailed_prices_with_cpi,
        cpi_monthly_avg,
        on='YearMonth',
        how='left'
    )
    detailed_prices_with_cpi['price_real'] = (detailed_prices_with_cpi['price'] / detailed_prices_with_cpi['cpi_value']) * base_cpi
    detailed_prices_with_cpi['price_real'] = detailed_prices_with_cpi['price_real'].fillna(detailed_prices_with_cpi['price'])
    detailed_prices_with_cpi['date'] = detailed_prices_with_cpi['date'].dt.strftime('%Y-%m-01')
    detailed_prices_with_cpi['month'] = detailed_prices_with_cpi['date'].astype(str).str[5:7].astype(int) # Extract month as int
    detailed_prices_with_cpi['year'] = detailed_prices_with_cpi['date'].astype(str).str[0:4].astype(int) # Extract year as int
    detailed_prices_with_cpi['price_ngn'] = detailed_prices_with_cpi['price']
    detailed_prices_with_cpi['unit'] = 'kg'
    detailed_prices_with_cpi['source'] = 'hfcp'
    detailed_prices_with_cpi['is_interpolated'] = 0
    prices_hfcp_df = detailed_prices_with_cpi[[
        'COMMODITY_SLUG',
        'MARKET_SLUG',
        'date',
        'month',
        'year',
        'price_ngn',
        'price_real',
        'unit',
        'source',
        'is_interpolated'
    ]].rename(columns={
        'COMMODITY_SLUG': 'commodity',
        'MARKET_SLUG': 'market'
    })

    # Prepare NBS data
    food_prices_with_cpi = all_cleaned_food_prices_df.copy()
    food_prices_with_cpi['YearMonth'] = food_prices_with_cpi['Date'].dt.to_period('M')
    food_prices_with_cpi = pd.merge(
        food_prices_with_cpi,
        cpi_monthly_avg,
        on='YearMonth',
        how='left'
    )
    food_prices_with_cpi['price_real'] = (food_prices_with_cpi['Price'] / food_prices_with_cpi['cpi_value']) * base_cpi
    food_prices_with_cpi['price_real'] = food_prices_with_cpi['price_real'].fillna(food_prices_with_cpi['Price'])
    food_prices_with_cpi['date'] = food_prices_with_cpi['Date'].dt.strftime('%Y-%m-01')
    food_prices_with_cpi['month'] = food_prices_with_cpi['Date'].dt.month
    food_prices_with_cpi['year'] = food_prices_with_cpi['Date'].dt.year
    food_prices_with_cpi['price_ngn'] = food_prices_with_cpi['Price']
    food_prices_with_cpi['unit'] = 'kg'
    food_prices_with_cpi['source'] = 'nbs'
    food_prices_with_cpi['is_interpolated'] = 0
    prices_nbs_df = food_prices_with_cpi[[
        'COMMODITY_SLUG',
        'MARKET_SLUG',
        'date',
        'month',
        'year',
        'price_ngn',
        'price_real',
        'unit',
        'source',
        'is_interpolated'
    ]].rename(columns={
        'COMMODITY_SLUG': 'commodity',
        'MARKET_SLUG': 'market'
    })

    final_prices_df = pd.concat([prices_hfcp_df, prices_nbs_df], ignore_index=True)
    final_prices_df_cleaned = final_prices_df.dropna(subset=['commodity', 'price_ngn', 'price_real'])
    final_prices_df_cleaned['date'] = pd.to_datetime(final_prices_df_cleaned['date'])
    return final_prices_df_cleaned

def generate_market_differentials(analysis_df, output_dir):
    """Calculates and saves market price differentials."""
    national_avg_prices = analysis_df.groupby(['Date', 'Item'])['price'].mean().reset_index()
    national_avg_prices.rename(columns={'price': 'national_avg_price'}, inplace=True)

    market_avg_prices = analysis_df.groupby(['Date', 'Item', 'MARKET_SLUG'])['price'].mean().reset_index()
    market_avg_prices.rename(columns={'price': 'market_avg_price'}, inplace=True)

    market_differentials = pd.merge(market_avg_prices,
                                    national_avg_prices,
                                    on=['Date', 'Item'],
                                    how='left')
    market_differentials['price_differential'] = market_differentials['market_avg_price'] - market_differentials['national_avg_price']

    market_differentials_final = market_differentials[[
        'Date', 'Item', 'MARKET_SLUG', 'market_avg_price', 'national_avg_price', 'price_differential'
    ]].copy()

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    market_differentials_final.to_csv(os.path.join(output_dir, 'market_differentials.csv'), index=False)
    print("Generated market_differentials.csv")

def generate_seasonal_patterns(analysis_df, output_dir):
    """Calculates and saves seasonal price patterns."""
    analysis_df['month'] = analysis_df['Date'].dt.month
    analysis_df['year'] = analysis_df['Date'].dt.year

    monthly_avg_prices = analysis_df.groupby(['Item', 'month'])['price'].mean().reset_index()
    monthly_avg_prices.rename(columns={'price': 'avg_monthly_price'}, inplace=True)

    commodity_annual_avg = analysis_df.groupby('Item')['price'].mean().reset_index()
    commodity_annual_avg.rename(columns={'price': 'overall_avg_price'}, inplace=True)

    seasonal_patterns = pd.merge(monthly_avg_prices,
                                 commodity_annual_avg,
                                 on='Item',
                                 how='left')
    seasonal_patterns['seasonal_index'] = (seasonal_patterns['avg_monthly_price'] / seasonal_patterns['overall_avg_price']) * 100

    seasonal_patterns_final = seasonal_patterns[[
        'Item', 'month', 'avg_monthly_price', 'seasonal_index'
    ]].copy()

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    seasonal_patterns_final.to_csv(os.path.join(output_dir, 'seasonal_patterns.csv'), index=False)
    print("Generated seasonal_patterns.csv")

def main(drive_path, output_dir, db_path):
    print("Starting data pipeline...")
    all_dataframes = load_dataframes_from_drive(drive_path)

    # Separate Excel and CSV dataframes for cleaning functions
    all_excel_dataframes = {k: v for k, v in all_dataframes.items() if k.endswith(('.xlsx', '.xls'))}
    # Note: all_dataframes already contains CSVs from the HFCP folder, so we don't need a separate filter for that here.

    all_cleaned_food_prices_df = clean_excel_dataframes(all_excel_dataframes)
    all_cleaned_detailed_prices_df = clean_detailed_csv_dataframes(all_dataframes)

    cpi_monthly_avg, base_cpi = prepare_cpi_data(all_dataframes)

    final_prices_df_cleaned = prepare_final_prices_df(all_cleaned_detailed_prices_df, all_cleaned_food_prices_df, cpi_monthly_avg, base_cpi)

    # Save to SQLite
    conn = sqlite3.connect(db_path)
    final_prices_df_cleaned.to_sql('prices', conn, if_exists='replace', index=False)
    conn.close()
    print(f"Data successfully saved to '{db_path}' in table 'prices'.")

    # Prepare analysis_df for market differentials and seasonal patterns
    analysis_df = all_cleaned_detailed_prices_df.copy()
    analysis_df['Date'] = pd.to_datetime(analysis_df['date']) # Ensure 'Date' is datetime
    analysis_df['price'] = pd.to_numeric(analysis_df['price'], errors='coerce')
    analysis_df.dropna(subset=['Date', 'price', 'commodity', 'MARKET_SLUG'], inplace=True)
    analysis_df.rename(columns={'commodity': 'Item'}, inplace=True)

    generate_market_differentials(analysis_df, output_dir)
    generate_seasonal_patterns(analysis_df, output_dir)

    print("Data pipeline finished.")

if __name__ == '__main__':
    drive_path = '/content/gdrive/My Drive/'
    output_dir = os.path.join(drive_path, 'data')
    db_path = os.path.join(drive_path, 'prices_clean.db')
    main(drive_path, output_dir, db_path)
```

I will now generate the `pipeline.py` script. This script will consolidate all the data loading, cleaning, and preprocessing logic developed so far. It will include:

```python
%%writefile "/content/gdrive/My Drive/pipeline.py"

import pandas as pd
import os
import glob
import re
import sqlite3

def load_dataframes_from_drive(drive_path):
    """Loads all CSV and Excel files from a specified Google Drive path."""
    all_dataframes = {}
    # Load CSVs
    csv_files = glob.glob(os.path.join(drive_path, '*.csv'))
    for file_path in csv_files:
        try:
            file_name = os.path.basename(file_path)
            all_dataframes[file_name] = pd.read_csv(file_path)
        except Exception as e:
            print(f"Error loading CSV {file_name}: {e}")

    # Load Excel files
    excel_files = glob.glob(os.path.join(drive_path, '*.xlsx')) + glob.glob(os.path.join(drive_path, '*.xls'))
    for file_path in excel_files:
        try:
            file_name = os.path.basename(file_path)
            all_dataframes[file_name] = pd.read_excel(file_path)
        except Exception as e:
            print(f"Error loading Excel {file_name}: {e}")

    # Load CSVs from NGA_2025_HFCP_v01_M_CSV folder
    matching_folders = glob.glob(os.path.join(drive_path, 'NGA_2025_HFCP*'))
    if matching_folders:
        folder_path = matching_folders[0]
        folder_csv_files = glob.glob(os.path.join(folder_path, '*.csv'))
        for file_path in folder_csv_files:
            try:
                file_name = os.path.basename(file_path)
                all_dataframes[file_name] = pd.read_csv(file_path)
            except Exception as e:
                print(f"Error loading CSV from folder {file_name}: {e}")

    return all_dataframes

def clean_excel_dataframes(all_excel_dataframes):
    """Cleans and consolidates Excel dataframes."""
    cleaned_excel_dataframes = {}
    for file_name, df in all_excel_dataframes.items():
        temp_df = df.copy()

        if not temp_df.empty and len(temp_df) > 0:
            if (isinstance(temp_df.columns[0], str) and temp_df.columns[0].startswith('Unnamed:')) or \
               (isinstance(temp_df.iloc[0, 0], str) and temp_df.iloc[0, 0].strip() in ['Items Label', 'Items', 'ItemLabels', 'NORTH CENTRAL'] and temp_df.columns[0] != temp_df.iloc[0,0]):
                temp_df.columns = temp_df.iloc[0]
                temp_df = temp_df[1:].reset_index(drop=True)

            temp_df = temp_df.loc[:, temp_df.columns.notna()]
            temp_df.columns = temp_df.columns.astype(str).str.strip()

            cols = []
            seen = {}
            for item in temp_df.columns:
                original_name = str(item).strip()
                name = original_name
                count = 0
                while name in seen:
                    count += 1
                    name = f"{original_name}_{count}"
                cols.append(name)
                seen[name] = True
            temp_df.columns = cols

        if 'Items Label' in temp_df.columns:
            temp_df = temp_df[temp_df['Items Label'].notna()]
            temp_df = temp_df[temp_df['Items Label'] != 'NORTH CENTRAL']
            temp_df = temp_df.dropna(how='all')
        elif 'Items' in temp_df.columns:
            temp_df = temp_df[temp_df['Items'].notna()]
            temp_df = temp_df[temp_df['Items'] != 'NORTH CENTRAL']
            temp_df = temp_df.dropna(how='all')

        price_columns = [col for col in temp_df.columns if col.startswith('Average of')]
        id_vars = [col for col in temp_df.columns if col not in price_columns]
        id_vars = list(dict.fromkeys(id_vars))
        id_vars = [col for col in id_vars if col in temp_df.columns]

        if price_columns and not temp_df.empty:
            melted_df = pd.melt(temp_df, id_vars=id_vars, value_vars=price_columns, var_name='Month_Year_Avg', value_name='Price')
            melted_df['Date_Str'] = melted_df['Month_Year_Avg'].str.replace('Average of ', '', regex=False)
            melted_df['Date'] = pd.to_datetime(melted_df['Date_Str'], format='%b-%y', errors='coerce')
            melted_df['Date'] = melted_df['Date'].fillna(pd.to_datetime(melted_df['Date_Str'], format='%b-%Y', errors='coerce'))
            melted_df = melted_df.dropna(subset=['Date'])
            melted_df['Price'] = pd.to_numeric(melted_df['Price'], errors='coerce')
            melted_df = melted_df.dropna(subset=['Price'])
            melted_df = melted_df.drop(columns=['Month_Year_Avg', 'Date_Str'], errors='ignore')
            cleaned_excel_dataframes[file_name] = melted_df

    all_cleaned_food_prices_df = pd.concat(cleaned_excel_dataframes.values(), ignore_index=True)

    all_cleaned_food_prices_df['Item'] = all_cleaned_food_prices_df['Items Label'].fillna(
        all_cleaned_food_prices_df['Items']).fillna(
        all_cleaned_food_prices_df['ItemLabel']).fillna(
        all_cleaned_food_prices_df['ITEMS']).fillna(
        all_cleaned_food_prices_df['Items_1']
    )
    all_cleaned_food_prices_df['MoM'] = pd.to_numeric(all_cleaned_food_prices_df['MoM'], errors='coerce')
    all_cleaned_food_prices_df['YoY'] = pd.to_numeric(all_cleaned_food_prices_df['YoY'], errors='coerce')

    def extract_price_location(text):
        if pd.isna(text):
            return None, None
        text = str(text).strip()
        match = re.search(r'(.+)\s*\(([-+]?\d*\.?\d+)\)', text)
        if match:
            location = match.group(1).strip()
            price = float(match.group(2))
            return price, location
        try:
            price = float(text)
            return price, None
        except ValueError:
            return None, text

    all_cleaned_food_prices_df[['Highest_Price', 'Highest_Location']] = all_cleaned_food_prices_df['Highest'].apply(lambda x: pd.Series(extract_price_location(x)))
    all_cleaned_food_prices_df[['Lowest_Price', 'Lowest_Location']] = all_cleaned_food_prices_df['Lowest'].apply(lambda x: pd.Series(extract_price_location(x)))

    columns_to_drop = [
        'Items Label', 'Items', 'ItemLabel', 'ITEMS', 'Items_1',
        'Highest', 'Lowest', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Grand Total'
    ]
    regional_cols = ['NORTH CENTRAL', 'NORTH EAST', 'NORTH WEST', 'SOUTH EAST', 'SOUTH SOUTH', 'SOUTH WEST']
    all_cleaned_food_prices_df = all_cleaned_food_prices_df.drop(
        columns=[col for col in columns_to_drop + regional_cols if col in all_cleaned_food_prices_df.columns]
    )

    all_cleaned_food_prices_df['Highest_Price'] = all_cleaned_food_prices_df['Highest_Price'].fillna(all_cleaned_food_prices_df['Price'])
    all_cleaned_food_prices_df['Lowest_Price'] = all_cleaned_food_prices_df['Lowest_Price'].fillna(all_cleaned_food_prices_df['Price'])

    all_cleaned_food_prices_df['COMMODITY_SLUG'] = all_cleaned_food_prices_df['Item'].str.lower().str.replace(' ', '_')
    all_cleaned_food_prices_df['MARKET_SLUG'] = 'nigeria_national_market'

    return all_cleaned_food_prices_df

def clean_detailed_csv_dataframes(all_dataframes):
    """Cleans and consolidates detailed CSV dataframes from HFCP folder."""
    cleaned_detailed_csv_dataframes = {}
    hfcp_csv_files = [k for k in all_dataframes.keys() if k.startswith('level2_price_') or k == 'Raw_data_spatialrev.csv']

    for file_name in hfcp_csv_files:
        df = all_dataframes[file_name].copy()
        df.columns = df.columns.str.lower().str.replace(' ', '_')

        if 'date' in df.columns:
            df['date'] = pd.to_datetime(df['date'], errors='coerce')
            df = df.dropna(subset=['date'])

        for col in ['price', 'daily_mean', 'daily_sd']:
            if col in df.columns:
                df[col] = pd.to_numeric(df[col], errors='coerce')

        subset_for_dropna = []
        if 'price' in df.columns:
            subset_for_dropna.append('price')
        if 'commodity' in df.columns:
            subset_for_dropna.append('commodity')

        if subset_for_dropna:
            df = df.dropna(subset=subset_for_dropna)

        # Create COMMODITY_SLUG for all_cleaned_detailed_prices_df
        df['COMMODITY_SLUG'] = df['commodity'].str.lower().str.replace(' ', '_')

        # Create MARKET_SLUG for all_cleaned_detailed_prices_df
        def create_market_slug(row):
            components = []
            if pd.notna(row['geopol_zon']):
                components.append(str(row['geopol_zon']))
            if pd.notna(row['statelabel']):
                components.append(str(row['statelabel']))
            if pd.notna(row['lgalabel']):
                components.append(str(row['lgalabel']))
            slug = '_'.join(components).lower()
            slug = re.sub(r'[^a-z0-9_]', '', slug)
            return slug

        df['MARKET_SLUG'] = df.apply(create_market_slug, axis=1)

        cleaned_detailed_csv_dataframes[file_name] = df

    all_cleaned_detailed_prices_df = pd.concat(cleaned_detailed_csv_dataframes.values(), ignore_index=True)
    return all_cleaned_detailed_prices_df

def prepare_cpi_data(all_dataframes):
    """Prepares CPI data for real price calculations."""
    cpi_df = all_dataframes['consumer-price-indices_nga.csv'].copy()
    cpi_df['StartDate'] = pd.to_datetime(cpi_df['StartDate'], errors='coerce')
    cpi_df['Value'] = pd.to_numeric(cpi_df['Value'], errors='coerce')
    cpi_df = cpi_df.dropna(subset=['StartDate', 'Value'])
    cpi_df = cpi_df.sort_values(by='StartDate').reset_index(drop=True)

    cpi_filtered_df = cpi_df[cpi_df['Item'] == 'Consumer Prices, General Indices (2015 = 100)'].copy()
    cpi_filtered_df['YearMonth'] = cpi_filtered_df['StartDate'].dt.to_period('M')
    cpi_monthly_avg = cpi_filtered_df.groupby('YearMonth')['Value'].mean().reset_index()
    cpi_monthly_avg.rename(columns={'Value': 'cpi_value'}, inplace=True)

    cpi_2024 = cpi_monthly_avg[cpi_monthly_avg['YearMonth'].dt.year == 2024]['cpi_value']
    if not cpi_2024.empty:
        base_cpi = cpi_2024.mean()
    else:
        base_cpi = cpi_monthly_avg['cpi_value'].mean()

    return cpi_monthly_avg, base_cpi

def prepare_final_prices_df(all_cleaned_detailed_prices_df, all_cleaned_food_prices_df, cpi_monthly_avg, base_cpi):
    """Prepares and consolidates all price data into a single DataFrame."""
    # Prepare HFCP data
    detailed_prices_with_cpi = all_cleaned_detailed_prices_df.copy()
    detailed_prices_with_cpi['YearMonth'] = detailed_prices_with_cpi['date'].dt.to_period('M')
    detailed_prices_with_cpi = pd.merge(
        detailed_prices_with_cpi,
        cpi_monthly_avg,
        on='YearMonth',
        how='left'
    )
    detailed_prices_with_cpi['price_real'] = (detailed_prices_with_cpi['price'] / detailed_prices_with_cpi['cpi_value']) * base_cpi
    detailed_prices_with_cpi['price_real'] = detailed_prices_with_cpi['price_real'].fillna(detailed_prices_with_cpi['price'])
    detailed_prices_with_cpi['date'] = detailed_prices_with_cpi['date'].dt.strftime('%Y-%m-01')
    detailed_prices_with_cpi['month'] = detailed_prices_with_cpi['date'].astype(str).str[5:7].astype(int) # Extract month as int
    detailed_prices_with_cpi['year'] = detailed_prices_with_cpi['date'].astype(str).str[0:4].astype(int) # Extract year as int
    detailed_prices_with_cpi['price_ngn'] = detailed_prices_with_cpi['price']
    detailed_prices_with_cpi['unit'] = 'kg'
    detailed_prices_with_cpi['source'] = 'hfcp'
    detailed_prices_with_cpi['is_interpolated'] = 0
    prices_hfcp_df = detailed_prices_with_cpi[[
        'COMMODITY_SLUG',
        'MARKET_SLUG',
        'date',
        'month',
        'year',
        'price_ngn',
        'price_real',
        'unit',
        'source',
        'is_interpolated'
    ]].rename(columns={
        'COMMODITY_SLUG': 'commodity',
        'MARKET_SLUG': 'market'
    })

    # Prepare NBS data
    food_prices_with_cpi = all_cleaned_food_prices_df.copy()
    food_prices_with_cpi['YearMonth'] = food_prices_with_cpi['Date'].dt.to_period('M')
    food_prices_with_cpi = pd.merge(
        food_prices_with_cpi,
        cpi_monthly_avg,
        on='YearMonth',
        how='left'
    )
    food_prices_with_cpi['price_real'] = (food_prices_with_cpi['Price'] / food_prices_with_cpi['cpi_value']) * base_cpi
    food_prices_with_cpi['price_real'] = food_prices_with_cpi['price_real'].fillna(food_prices_with_cpi['Price'])
    food_prices_with_cpi['date'] = food_prices_with_cpi['Date'].dt.strftime('%Y-%m-01')
    food_prices_with_cpi['month'] = food_prices_with_cpi['Date'].dt.month
    food_prices_with_cpi['year'] = food_prices_with_cpi['Date'].dt.year
    food_prices_with_cpi['price_ngn'] = food_prices_with_cpi['Price']
    food_prices_with_cpi['unit'] = 'kg'
    food_prices_with_cpi['source'] = 'nbs'
    food_prices_with_cpi['is_interpolated'] = 0
    prices_nbs_df = food_prices_with_cpi[[
        'COMMODITY_SLUG',
        'MARKET_SLUG',
        'date',
        'month',
        'year',
        'price_ngn',
        'price_real',
        'unit',
        'source',
        'is_interpolated'
    ]].rename(columns={
        'COMMODITY_SLUG': 'commodity',
        'MARKET_SLUG': 'market'
    })

    final_prices_df = pd.concat([prices_hfcp_df, prices_nbs_df], ignore_index=True)
    final_prices_df_cleaned = final_prices_df.dropna(subset=['commodity', 'price_ngn', 'price_real'])
    final_prices_df_cleaned['date'] = pd.to_datetime(final_prices_df_cleaned['date'])
    return final_prices_df_cleaned

def generate_market_differentials(analysis_df, output_dir):
    """Calculates and saves market price differentials."""
    national_avg_prices = analysis_df.groupby(['Date', 'Item'])['price'].mean().reset_index()
    national_avg_prices.rename(columns={'price': 'national_avg_price'}, inplace=True)

    market_avg_prices = analysis_df.groupby(['Date', 'Item', 'MARKET_SLUG'])['price'].mean().reset_index()
    market_avg_prices.rename(columns={'price': 'market_avg_price'}, inplace=True)

    market_differentials = pd.merge(market_avg_prices,
                                    national_avg_prices,
                                    on=['Date', 'Item'],
                                    how='left')
    market_differentials['price_differential'] = market_differentials['market_avg_price'] - market_differentials['national_avg_price']

    market_differentials_final = market_differentials[[
        'Date', 'Item', 'MARKET_SLUG', 'market_avg_price', 'national_avg_price', 'price_differential'
    ]].copy()

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    market_differentials_final.to_csv(os.path.join(output_dir, 'market_differentials.csv'), index=False)
    print("Generated market_differentials.csv")

def generate_seasonal_patterns(analysis_df, output_dir):
    """Calculates and saves seasonal price patterns."""
    analysis_df['month'] = analysis_df['Date'].dt.month
    analysis_df['year'] = analysis_df['Date'].dt.year

    monthly_avg_prices = analysis_df.groupby(['Item', 'month'])['price'].mean().reset_index()
    monthly_avg_prices.rename(columns={'price': 'avg_monthly_price'}, inplace=True)

    commodity_annual_avg = analysis_df.groupby('Item')['price'].mean().reset_index()
    commodity_annual_avg.rename(columns={'price': 'overall_avg_price'}, inplace=True)

    seasonal_patterns = pd.merge(monthly_avg_prices,
                                 commodity_annual_avg,
                                 on='Item',
                                 how='left')
    seasonal_patterns['seasonal_index'] = (seasonal_patterns['avg_monthly_price'] / seasonal_patterns['overall_avg_price']) * 100

    seasonal_patterns_final = seasonal_patterns[[
        'Item', 'month', 'avg_monthly_price', 'seasonal_index'
    ]].copy()

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    seasonal_patterns_final.to_csv(os.path.join(output_dir, 'seasonal_patterns.csv'), index=False)
    print("Generated seasonal_patterns.csv")

def main(drive_path, output_dir, db_path):
    print("Starting data pipeline...")
    all_dataframes = load_dataframes_from_drive(drive_path)

    # Separate Excel and CSV dataframes for cleaning functions
    all_excel_dataframes = {k: v for k, v in all_dataframes.items() if k.endswith(('.xlsx', '.xls'))}
    # Note: all_dataframes already contains CSVs from the HFCP folder, so we don't need a separate filter for that here.

    all_cleaned_food_prices_df = clean_excel_dataframes(all_excel_dataframes)
    all_cleaned_detailed_prices_df = clean_detailed_csv_dataframes(all_dataframes)

    cpi_monthly_avg, base_cpi = prepare_cpi_data(all_dataframes)

    final_prices_df_cleaned = prepare_final_prices_df(all_cleaned_detailed_prices_df, all_cleaned_food_prices_df, cpi_monthly_avg, base_cpi)

    # Save to SQLite
    conn = sqlite3.connect(db_path)
    final_prices_df_cleaned.to_sql('prices', conn, if_exists='replace', index=False)
    conn.close()
    print(f"Data successfully saved to '{db_path}' in table 'prices'.")

    # Prepare analysis_df for market differentials and seasonal patterns
    analysis_df = all_cleaned_detailed_prices_df.copy()
    analysis_df['Date'] = pd.to_datetime(analysis_df['date']) # Ensure 'Date' is datetime
    analysis_df['price'] = pd.to_numeric(analysis_df['price'], errors='coerce')
    analysis_df.dropna(subset=['Date', 'price', 'commodity', 'MARKET_SLUG'], inplace=True)
    analysis_df.rename(columns={'commodity': 'Item'}, inplace=True)

    generate_market_differentials(analysis_df, output_dir)
    generate_seasonal_patterns(analysis_df, output_dir)

    print("Data pipeline finished.")

if __name__ == '__main__':
    drive_path = '/content/gdrive/My Drive/'
    output_dir = os.path.join(drive_path, 'data')
    db_path = os.path.join(drive_path, 'prices_clean.db')
    main(drive_path, output_dir, db_path)
```

```python
%%writefile "/content/gdrive/My Drive/pipeline.py"

import pandas as pd
import os
import glob
import re
import sqlite3

def load_dataframes_from_drive(drive_path):
    """Loads all CSV and Excel files from a specified Google Drive path."""
    all_dataframes = {}
    # Load CSVs
    csv_files = glob.glob(os.path.join(drive_path, '*.csv'))
    for file_path in csv_files:
        try:
            file_name = os.path.basename(file_path)
            all_dataframes[file_name] = pd.read_csv(file_path)
        except Exception as e:
            print(f"Error loading CSV {file_name}: {e}")

    # Load Excel files
    excel_files = glob.glob(os.path.join(drive_path, '*.xlsx')) + glob.glob(os.path.join(drive_path, '*.xls'))
    for file_path in excel_files:
        try:
            file_name = os.path.basename(file_path)
            all_dataframes[file_name] = pd.read_excel(file_path)
        except Exception as e:
            print(f"Error loading Excel {file_name}: {e}")

    # Load CSVs from NGA_2025_HFCP_v01_M_CSV folder
    matching_folders = glob.glob(os.path.join(drive_path, 'NGA_2025_HFCP*'))
    if matching_folders:
        folder_path = matching_folders[0]
        folder_csv_files = glob.glob(os.path.join(folder_path, '*.csv'))
        for file_path in folder_csv_files:
            try:
                file_name = os.path.basename(file_path)
                all_dataframes[file_name] = pd.read_csv(file_path)
            except Exception as e:
                print(f"Error loading CSV from folder {file_name}: {e}")

    return all_dataframes

def clean_excel_dataframes(all_excel_dataframes):
    """Cleans and consolidates Excel dataframes."""
    cleaned_excel_dataframes = {}
    for file_name, df in all_excel_dataframes.items():
        temp_df = df.copy()

        if not temp_df.empty and len(temp_df) > 0:
            if (isinstance(temp_df.columns[0], str) and temp_df.columns[0].startswith('Unnamed:')) or \
               (isinstance(temp_df.iloc[0, 0], str) and temp_df.iloc[0, 0].strip() in ['Items Label', 'Items', 'ItemLabels', 'NORTH CENTRAL'] and temp_df.columns[0] != temp_df.iloc[0,0]):
                temp_df.columns = temp_df.iloc[0]
                temp_df = temp_df[1:].reset_index(drop=True)

            temp_df = temp_df.loc[:, temp_df.columns.notna()]
            temp_df.columns = temp_df.columns.astype(str).str.strip()

            cols = []
            seen = {}
            for item in temp_df.columns:
                original_name = str(item).strip()
                name = original_name
                count = 0
                while name in seen:
                    count += 1
                    name = f"{original_name}_{count}"
                cols.append(name)
                seen[name] = True
            temp_df.columns = cols

        if 'Items Label' in temp_df.columns:
            temp_df = temp_df[temp_df['Items Label'].notna()]
            temp_df = temp_df[temp_df['Items Label'] != 'NORTH CENTRAL']
            temp_df = temp_df.dropna(how='all')
        elif 'Items' in temp_df.columns:
            temp_df = temp_df[temp_df['Items'].notna()]
            temp_df = temp_df[temp_df['Items'] != 'NORTH CENTRAL']
            temp_df = temp_df.dropna(how='all')

        price_columns = [col for col in temp_df.columns if col.startswith('Average of')]
        id_vars = [col for col in temp_df.columns if col not in price_columns]
        id_vars = list(dict.fromkeys(id_vars))
        id_vars = [col for col in id_vars if col in temp_df.columns]

        if price_columns and not temp_df.empty:
            melted_df = pd.melt(temp_df, id_vars=id_vars, value_vars=price_columns, var_name='Month_Year_Avg', value_name='Price')
            melted_df['Date_Str'] = melted_df['Month_Year_Avg'].str.replace('Average of ', '', regex=False)
            melted_df['Date'] = pd.to_datetime(melted_df['Date_Str'], format='%b-%y', errors='coerce')
            melted_df['Date'] = melted_df['Date'].fillna(pd.to_datetime(melted_df['Date_Str'], format='%b-%Y', errors='coerce'))
            melted_df = melted_df.dropna(subset=['Date'])
            melted_df['Price'] = pd.to_numeric(melted_df['Price'], errors='coerce')
            melted_df = melted_df.dropna(subset=['Price'])
            melted_df = melted_df.drop(columns=['Month_Year_Avg', 'Date_Str'], errors='ignore')
            cleaned_excel_dataframes[file_name] = melted_df

    all_cleaned_food_prices_df = pd.concat(cleaned_excel_dataframes.values(), ignore_index=True)

    all_cleaned_food_prices_df['Item'] = all_cleaned_food_prices_df['Items Label'].fillna(
        all_cleaned_food_prices_df['Items']).fillna(
        all_cleaned_food_prices_df['ItemLabel']).fillna(
        all_cleaned_food_prices_df['ITEMS']).fillna(
        all_cleaned_food_prices_df['Items_1']
    )
    all_cleaned_food_prices_df['MoM'] = pd.to_numeric(all_cleaned_food_prices_df['MoM'], errors='coerce')
    all_cleaned_food_prices_df['YoY'] = pd.to_numeric(all_cleaned_food_prices_df['YoY'], errors='coerce')

    def extract_price_location(text):
        if pd.isna(text):
            return None, None
        text = str(text).strip()
        match = re.search(r'(.+)\s*\(([-+]?\d*\.?\d+)\)', text)
        if match:
            location = match.group(1).strip()
            price = float(match.group(2))
            return price, location
        try:
            price = float(text)
            return price, None
        except ValueError:
            return None, text

    all_cleaned_food_prices_df[['Highest_Price', 'Highest_Location']] = all_cleaned_food_prices_df['Highest'].apply(lambda x: pd.Series(extract_price_location(x)))
    all_cleaned_food_prices_df[['Lowest_Price', 'Lowest_Location']] = all_cleaned_food_prices_df['Lowest'].apply(lambda x: pd.Series(extract_price_location(x)))

    columns_to_drop = [
        'Items Label', 'Items', 'ItemLabel', 'ITEMS', 'Items_1',
        'Highest', 'Lowest', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Grand Total'
    ]
    regional_cols = ['NORTH CENTRAL', 'NORTH EAST', 'NORTH WEST', 'SOUTH EAST', 'SOUTH SOUTH', 'SOUTH WEST']
    all_cleaned_food_prices_df = all_cleaned_food_prices_df.drop(
        columns=[col for col in columns_to_drop + regional_cols if col in all_cleaned_food_prices_df.columns]
    )

    all_cleaned_food_prices_df['Highest_Price'] = all_cleaned_food_prices_df['Highest_Price'].fillna(all_cleaned_food_prices_df['Price'])
    all_cleaned_food_prices_df['Lowest_Price'] = all_cleaned_food_prices_df['Lowest_Price'].fillna(all_cleaned_food_prices_df['Price'])

    all_cleaned_food_prices_df['COMMODITY_SLUG'] = all_cleaned_food_prices_df['Item'].str.lower().str.replace(' ', '_')
    all_cleaned_food_prices_df['MARKET_SLUG'] = 'nigeria_national_market'

    return all_cleaned_food_prices_df

def clean_detailed_csv_dataframes(all_dataframes):
    """Cleans and consolidates detailed CSV dataframes from HFCP folder."""
    cleaned_detailed_csv_dataframes = {}
    hfcp_csv_files = [k for k in all_dataframes.keys() if k.startswith('level2_price_') or k == 'Raw_data_spatialrev.csv']

    for file_name in hfcp_csv_files:
        df = all_dataframes[file_name].copy()
        df.columns = df.columns.str.lower().str.replace(' ', '_')

        if 'date' in df.columns:
            df['date'] = pd.to_datetime(df['date'], errors='coerce')
            df = df.dropna(subset=['date'])

        for col in ['price', 'daily_mean', 'daily_sd']:
            if col in df.columns:
                df[col] = pd.to_numeric(df[col], errors='coerce')

        subset_for_dropna = []
        if 'price' in df.columns:
            subset_for_dropna.append('price')
        if 'commodity' in df.columns:
            subset_for_dropna.append('commodity')

        if subset_for_dropna:
            df = df.dropna(subset=subset_for_dropna)

        # Create COMMODITY_SLUG for all_cleaned_detailed_prices_df
        df['COMMODITY_SLUG'] = df['commodity'].str.lower().str.replace(' ', '_')

        # Create MARKET_SLUG for all_cleaned_detailed_prices_df
        def create_market_slug(row):
            components = []
            if pd.notna(row['geopol_zon']):
                components.append(str(row['geopol_zon']))
            if pd.notna(row['statelabel']):
                components.append(str(row['statelabel']))
            if pd.notna(row['lgalabel']):
                components.append(str(row['lgalabel']))
            slug = '_'.join(components).lower()
            slug = re.sub(r'[^a-z0-9_]', '', slug)
            return slug

        df['MARKET_SLUG'] = df.apply(create_market_slug, axis=1)

        cleaned_detailed_csv_dataframes[file_name] = df

    all_cleaned_detailed_prices_df = pd.concat(cleaned_detailed_csv_dataframes.values(), ignore_index=True)
    return all_cleaned_detailed_prices_df

def prepare_cpi_data(all_dataframes):
    """Prepares CPI data for real price calculations."""
    cpi_df = all_dataframes['consumer-price-indices_nga.csv'].copy()
    cpi_df['StartDate'] = pd.to_datetime(cpi_df['StartDate'], errors='coerce')
    cpi_df['Value'] = pd.to_numeric(cpi_df['Value'], errors='coerce')
    cpi_df = cpi_df.dropna(subset=['StartDate', 'Value'])
    cpi_df = cpi_df.sort_values(by='StartDate').reset_index(drop=True)

    cpi_filtered_df = cpi_df[cpi_df['Item'] == 'Consumer Prices, General Indices (2015 = 100)'].copy()
    cpi_filtered_df['YearMonth'] = cpi_filtered_df['StartDate'].dt.to_period('M')
    cpi_monthly_avg = cpi_filtered_df.groupby('YearMonth')['Value'].mean().reset_index()
    cpi_monthly_avg.rename(columns={'Value': 'cpi_value'}, inplace=True)

    cpi_2024 = cpi_monthly_avg[cpi_monthly_avg['YearMonth'].dt.year == 2024]['cpi_value']
    if not cpi_2024.empty:
        base_cpi = cpi_2024.mean()
    else:
        base_cpi = cpi_monthly_avg['cpi_value'].mean()

    return cpi_monthly_avg, base_cpi

def prepare_final_prices_df(all_cleaned_detailed_prices_df, all_cleaned_food_prices_df, cpi_monthly_avg, base_cpi):
    """Prepares and consolidates all price data into a single DataFrame."""
    # Prepare HFCP data
    detailed_prices_with_cpi = all_cleaned_detailed_prices_df.copy()
    detailed_prices_with_cpi['YearMonth'] = detailed_prices_with_cpi['date'].dt.to_period('M')
    detailed_prices_with_cpi = pd.merge(
        detailed_prices_with_cpi,
        cpi_monthly_avg,
        on='YearMonth',
        how='left'
    )
    detailed_prices_with_cpi['price_real'] = (detailed_prices_with_cpi['price'] / detailed_prices_with_cpi['cpi_value']) * base_cpi
    detailed_prices_with_cpi['price_real'] = detailed_prices_with_cpi['price_real'].fillna(detailed_prices_with_cpi['price'])
    detailed_prices_with_cpi['date'] = detailed_prices_with_cpi['date'].dt.strftime('%Y-%m-01')
    detailed_prices_with_cpi['month'] = detailed_prices_with_cpi['date'].astype(str).str[5:7].astype(int) # Extract month as int
    detailed_prices_with_cpi['year'] = detailed_prices_with_cpi['date'].astype(str).str[0:4].astype(int) # Extract year as int
    detailed_prices_with_cpi['price_ngn'] = detailed_prices_with_cpi['price']
    detailed_prices_with_cpi['unit'] = 'kg'
    detailed_prices_with_cpi['source'] = 'hfcp'
    detailed_prices_with_cpi['is_interpolated'] = 0
    prices_hfcp_df = detailed_prices_with_cpi[[
        'COMMODITY_SLUG',
        'MARKET_SLUG',
        'date',
        'month',
        'year',
        'price_ngn',
        'price_real',
        'unit',
        'source',
        'is_interpolated'
    ]].rename(columns={
        'COMMODITY_SLUG': 'commodity',
        'MARKET_SLUG': 'market'
    })

    # Prepare NBS data
    food_prices_with_cpi = all_cleaned_food_prices_df.copy()
    food_prices_with_cpi['YearMonth'] = food_prices_with_cpi['Date'].dt.to_period('M')
    food_prices_with_cpi = pd.merge(
        food_prices_with_cpi,
        cpi_monthly_avg,
        on='YearMonth',
        how='left'
    )
    food_prices_with_cpi['price_real'] = (food_prices_with_cpi['Price'] / food_prices_with_cpi['cpi_value']) * base_cpi
    food_prices_with_cpi['price_real'] = food_prices_with_cpi['price_real'].fillna(food_prices_with_cpi['Price'])
    food_prices_with_cpi['date'] = food_prices_with_cpi['Date'].dt.strftime('%Y-%m-01')
    food_prices_with_cpi['month'] = food_prices_with_cpi['Date'].dt.month
    food_prices_with_cpi['year'] = food_prices_with_cpi['Date'].dt.year
    food_prices_with_cpi['price_ngn'] = food_prices_with_cpi['Price']
    food_prices_with_cpi['unit'] = 'kg'
    food_prices_with_cpi['source'] = 'nbs'
    food_prices_with_cpi['is_interpolated'] = 0
    prices_nbs_df = food_prices_with_cpi[[
        'COMMODITY_SLUG',
        'MARKET_SLUG',
        'date',
        'month',
        'year',
        'price_ngn',
        'price_real',
        'unit',
        'source',
        'is_interpolated'
    ]].rename(columns={
        'COMMODITY_SLUG': 'commodity',
        'MARKET_SLUG': 'market'
    })

    final_prices_df = pd.concat([prices_hfcp_df, prices_nbs_df], ignore_index=True)
    final_prices_df_cleaned = final_prices_df.dropna(subset=['commodity', 'price_ngn', 'price_real'])
    final_prices_df_cleaned['date'] = pd.to_datetime(final_prices_df_cleaned['date'])
    return final_prices_df_cleaned

def generate_market_differentials(analysis_df, output_dir):
    """Calculates and saves market price differentials."""
    national_avg_prices = analysis_df.groupby(['Date', 'Item'])['price'].mean().reset_index()
    national_avg_prices.rename(columns={'price': 'national_avg_price'}, inplace=True)

    market_avg_prices = analysis_df.groupby(['Date', 'Item', 'MARKET_SLUG'])['price'].mean().reset_index()
    market_avg_prices.rename(columns={'price': 'market_avg_price'}, inplace=True)

    market_differentials = pd.merge(market_avg_prices,
                                    national_avg_prices,
                                    on=['Date', 'Item'],
                                    how='left')
    market_differentials['price_differential'] = market_differentials['market_avg_price'] - market_differentials['national_avg_price']

    market_differentials_final = market_differentials[[
        'Date', 'Item', 'MARKET_SLUG', 'market_avg_price', 'national_avg_price', 'price_differential'
    ]].copy()

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    market_differentials_final.to_csv(os.path.join(output_dir, 'market_differentials.csv'), index=False)
    print("Generated market_differentials.csv")

def generate_seasonal_patterns(analysis_df, output_dir):
    """Calculates and saves seasonal price patterns."""
    analysis_df['month'] = analysis_df['Date'].dt.month
    analysis_df['year'] = analysis_df['Date'].dt.year

    monthly_avg_prices = analysis_df.groupby(['Item', 'month'])['price'].mean().reset_index()
    monthly_avg_prices.rename(columns={'price': 'avg_monthly_price'}, inplace=True)

    commodity_annual_avg = analysis_df.groupby('Item')['price'].mean().reset_index()
    commodity_annual_avg.rename(columns={'price': 'overall_avg_price'}, inplace=True)

    seasonal_patterns = pd.merge(monthly_avg_prices,
                                 commodity_annual_avg,
                                 on='Item',
                                 how='left')
    seasonal_patterns['seasonal_index'] = (seasonal_patterns['avg_monthly_price'] / seasonal_patterns['overall_avg_price']) * 100

    seasonal_patterns_final = seasonal_patterns[[
        'Item', 'month', 'avg_monthly_price', 'seasonal_index'
    ]].copy()

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    seasonal_patterns_final.to_csv(os.path.join(output_dir, 'seasonal_patterns.csv'), index=False)
    print("Generated seasonal_patterns.csv")

def main(drive_path, output_dir, db_path):
    print("Starting data pipeline...")
    all_dataframes = load_dataframes_from_drive(drive_path)

    # Separate Excel and CSV dataframes for cleaning functions
    all_excel_dataframes = {k: v for k, v in all_dataframes.items() if k.endswith(('.xlsx', '.xls'))}
    # Note: all_dataframes already contains CSVs from the HFCP folder, so we don't need a separate filter for that here.

    all_cleaned_food_prices_df = clean_excel_dataframes(all_excel_dataframes)
    all_cleaned_detailed_prices_df = clean_detailed_csv_dataframes(all_dataframes)

    cpi_monthly_avg, base_cpi = prepare_cpi_data(all_dataframes)

    final_prices_df_cleaned = prepare_final_prices_df(all_cleaned_detailed_prices_df, all_cleaned_food_prices_df, cpi_monthly_avg, base_cpi)

    # Save to SQLite
    conn = sqlite3.connect(db_path)
    final_prices_df_cleaned.to_sql('prices', conn, if_exists='replace', index=False)
    conn.close()
    print(f"Data successfully saved to '{db_path}' in table 'prices'.")

    # Prepare analysis_df for market differentials and seasonal patterns
    analysis_df = all_cleaned_detailed_prices_df.copy()
    analysis_df['Date'] = pd.to_datetime(analysis_df['date']) # Ensure 'Date' is datetime
    analysis_df['price'] = pd.to_numeric(analysis_df['price'], errors='coerce')
    analysis_df.dropna(subset=['Date', 'price', 'commodity', 'MARKET_SLUG'], inplace=True)
    analysis_df.rename(columns={'commodity': 'Item'}, inplace=True)

    generate_market_differentials(analysis_df, output_dir)
    generate_seasonal_patterns(analysis_df, output_dir)

    print("Data pipeline finished.")

if __name__ == '__main__':
    drive_path = '/content/gdrive/My Drive/'
    output_dir = os.path.join(drive_path, 'data')
    db_path = os.path.join(drive_path, 'prices_clean.db')
    main(drive_path, output_dir, db_path)
```

## Project Documentation: AI-Powered Market Advisor Bot for Abuja - Data Pipeline

This document outlines the data preparation, cleaning, and integration pipeline developed for the AI-powered Market Advisor bot for Abuja. It covers the rationale behind key decisions and the processes implemented to transform raw data into a clean, structured format ready for analysis and modeling.

### 1. Project Goal
The primary objective of this project is to develop an AI-powered Market Advisor bot for Abuja. This bot aims to provide insights into food price dynamics, enabling better forecasting, and potentially informing bulk purchasing strategies or policy decisions related to food security. The current phase focuses on building a robust data foundation for this bot.

### 2. Data Sources & Ingestion
The pipeline integrates data from three primary sources, all stored on Google Drive and accessed via mounted drive:

*   **Consumer Price Index (CPI) Data (`consumer-price-indices_nga.csv`)**: Provides macroeconomic inflation indicators crucial for calculating real prices.
*   **High-Frequency Price Monitoring (HFCP) Data (from `NGA_2025_HFCP_v01_M_CSV` folder)**: Contains granular daily price data for various commodities across different markets in Nigeria.
*   **National Bureau of Statistics (NBS) Excel Files (`selected_food_*.xlsx`)**: Provides aggregated monthly food price data, often including month-over-month (MoM) and year-over-year (YoY) percentage changes, and information on highest/lowest prices and locations.

All these files are loaded into pandas DataFrames initially and stored in `all_dataframes` (for CSVs) and `all_excel_dataframes` (for Excel files) dictionaries.

### 3. Data Cleaning and Preprocessing

#### 3.1. NBS Excel Data (`all_cleaned_food_prices_df`)
This dataset required significant cleaning due to inconsistent formatting across multiple Excel files:

*   **Inconsistent Headers**: Many files had metadata rows before the actual header. A heuristic was applied to identify and promote the correct header row (e.g., checking for 'Unnamed:' columns or specific keywords like 'Items Label').
*   **Column Cleaning**: Column names were stripped of whitespace and made unique to avoid conflicts.
*   **Irrelevant Rows**: Rows containing summary data or headers like 'NORTH CENTRAL' (which indicated regional sub-tables) were identified and dropped.
*   **Melt Transformation**: The price data was in a wide format (e.g., 'Average of Jan-23', 'Average of Feb-23'). It was `melted` into a long format, creating a single `Price` column and a `Date` column from the month-year strings.
*   **Date Conversion**: Various date formats (`%b-%y`, `%b-%Y`) were handled during conversion to datetime objects, with unparseable dates being coerced to `NaT` and subsequently dropped.
*   **Price Conversion**: The `Price` column was converted to numeric, with errors coerced to `NaN` and then dropped.
*   **Item Consolidation**: Several columns (`Items Label`, `Items`, `ItemLabel`, `ITEMS`, `Items_1`) were used to identify commodities. These were consolidated into a single `Item` column to standardize commodity names.
*   **MoM/YoY Conversion**: 'MoM' and 'YoY' columns were converted to numeric types.
*   **Extraction from 'Highest'/'Lowest'**: The `Highest` and `Lowest` columns often contained both a price and a location (e.g., 'Location (1234.5)'). Regular expressions were used to extract `Highest_Price`, `Highest_Location`, `Lowest_Price`, and `Lowest_Location` into separate columns.
*   **Missing Price Imputation**: Missing values in `Highest_Price` and `Lowest_Price` were imputed with the `Price` from the same row, assuming that the recorded price serves as the best estimate when specific highest/lowest market data is unavailable for that period.
*   **Column Dropping**: Original, unconsolidated, and highly null columns (e.g., `Unnamed: 8`) were dropped.

#### 3.2. Detailed CSV Data (`all_cleaned_detailed_prices_df`)
This dataset, primarily from the HFCP source, contains daily granular price observations:

*   **Column Standardization**: All column names were converted to lowercase and spaces replaced with underscores for consistency.
*   **Date Conversion**: The `date` column was converted to datetime objects, with unparseable dates dropped.
*   **Numeric Conversion**: `price`, `daily_mean`, and `daily_sd` columns were converted to numeric, with errors coerced to `NaN`.
*   **Critical NaN Handling**: Rows with missing values in `price` or `commodity` were dropped, as these are essential for the analysis.

#### 3.3. CPI Data Preparation
Inflation data from `consumer-price-indices_nga.csv` was processed to enable real price calculations:

*   **Date and Value Conversion**: `StartDate` was converted to datetime and `Value` to numeric.
*   **Filtering**: The DataFrame was filtered to include only 'Consumer Prices, General Indices (2015 = 100)', which represents the overall inflation measure.
*   **Monthly Aggregation**: CPI values were grouped by `YearMonth` (created from `StartDate`) and the mean CPI value (`cpi_value`) was calculated for each month.
*   **Base CPI Calculation**: An average `base_cpi` was calculated using the mean `cpi_value` for 2024. If 2024 data was unavailable, the overall mean CPI was used as a fallback. This `base_cpi` serves as a reference point for inflation adjustments.

### 4. Data Standardization: Commodity and Market Slugs
To facilitate merging and consistent referencing, standardized slugs were created for commodities and markets:

*   **`COMMODITY_SLUG`**: Created for both `all_cleaned_detailed_prices_df` (from its `commodity` column) and `all_cleaned_food_prices_df` (from its `Item` column). This involves converting the item name to lowercase and replacing spaces with underscores.
*   **`MARKET_SLUG`**:
    *   For `all_cleaned_detailed_prices_df`, a `MARKET_SLUG` was created by concatenating and slugifying `geopol_zon`, `statelabel`, and `lgalabel` (e.g., `nbssouthsouth_bayelsa_kolokumaopokuma`).
    *   For `all_cleaned_food_prices_df`, as it represents aggregated national data, a general `nigeria_national_market` slug was assigned.

### 5. Data Integration and Real Price Calculation

#### 5.1. Handling Non-overlapping Data
Initial attempts to directly merge `all_cleaned_detailed_prices_df` and `all_cleaned_food_prices_df` failed due to significant differences in their 'Item' names and 'YearMonth' coverage. Given this, the decision was made to treat them as distinct, complementary data sources rather than forcing a direct merge. They are processed separately and then concatenated at the end.

#### 5.2. Calculating Real Prices
For both the HFCP-derived (`prices_hfcp_df`) and NBS-derived (`prices_nbs_df`) datasets, nominal prices were converted to real prices using the prepared CPI data:

*   **YearMonth Creation**: A `YearMonth` column was created in each price DataFrame.
*   **CPI Merge**: Each price DataFrame was left-merged with `cpi_monthly_avg` on the `YearMonth` column.
*   **`price_real` Calculation**: `price_real = (nominal_price / cpi_value) * base_cpi`. This adjusts the nominal price for inflation, expressing it in constant 2024 (or overall mean) Naira.
*   **Missing CPI Handling**: If a `cpi_value` was missing for a particular month, the `price_real` was filled with the original nominal price, implying no inflation adjustment could be made.

#### 5.3. Final Consolidation (`final_prices_df_cleaned`)

*   **Schema Alignment**: Both `prices_hfcp_df` and `prices_nbs_df` were transformed to align with a common `prices` table schema, including columns like `commodity`, `market`, `date`, `month`, `year`, `price_ngn` (nominal), `price_real`, `unit`, `source`, and `is_interpolated`.
*   **Concatenation**: The two DataFrames (`prices_hfcp_df` and `prices_nbs_df`) were concatenated into a single `final_prices_df`.
*   **Final NaN Drop**: Rows with missing values in critical columns (`commodity`, `price_ngn`, `price_real`) in the `final_prices_df` were dropped to ensure data integrity for downstream analysis, resulting in `final_prices_df_cleaned`.

### 6. Output and Deliverables

#### 6.1. SQLite Database (`prices_clean.db`)
The `final_prices_df_cleaned` DataFrame, containing all cleaned and consolidated price data, is saved to an SQLite database (`prices_clean.db`) in a table named `prices`. This provides a centralized, queryable, and easily accessible repository for the project's core price data.

#### 6.2. Market Analysis Files (CSV)
Two CSV files are generated to provide immediate insights:

*   **`market_differentials.csv`**: Calculated from `analysis_df` (which is derived from `all_cleaned_detailed_prices_df`). This file shows the difference between market-specific average prices and the national average price for each commodity per month. It helps identify regions where a commodity is consistently more expensive or cheaper.
*   **`seasonal_patterns.csv`**: Also calculated from `analysis_df`. This file captures recurring monthly price variations for each commodity by calculating average prices per month across all years and normalizing them into a `seasonal_index` (percentage deviation from the overall annual average price). This is vital for understanding and forecasting seasonal price fluctuations.

#### 6.3. Automation Script (`pipeline.py`)
A comprehensive `pipeline.py` script has been generated. This script encapsulates all the data loading, cleaning, preprocessing, real price calculation, market analysis (differentials and seasonality), and database integration steps. It allows for the entire data pipeline to be executed automatically, ensuring reproducibility and efficiency.

### 7. Key Decisions & Rationale

*   **Choice of `analysis_df`**: `all_cleaned_detailed_prices_df` (HFCP data) was specifically chosen for generating `market_differentials.csv` and `seasonal_patterns.csv` because it provides granular daily data across various specific markets, which is crucial for these types of analyses. The `wfp_food_prices_nga.csv` dataset, despite initial consideration, lacked a direct and clear date column, making it unsuitable for time-series and market differential analysis without further complex imputation or external data.
*   **Separate Data Sources**: Instead of forcing a problematic merge between HFCP and NBS data due to differing item names and time periods, the decision was made to treat them as distinct but complementary sources. They are cleaned and processed separately to their respective schemas and then concatenated, preserving the unique insights from each.
*   **Robust Cleaning for Excel Files**: The extensive cleaning steps for NBS Excel files (header handling, melting, item consolidation, extraction from `Highest`/`Lowest`) were necessary due to the high variability and complexity of their raw format. This ensured that valuable aggregated data was not lost.
*   **Imputation Strategy**: For missing `Highest_Price` and `Lowest_Price`, imputing with the `Price` from the same row is a pragmatic solution. It acknowledges that if specific extreme prices aren't reported, the observed price is the best available proxy, preventing data loss.
*   **Dropping Critical NaNs**: Removing rows with missing `commodity`, `price_ngn`, or `price_real` in the final consolidated DataFrame (`final_prices_df_cleaned`) is a critical step to maintain data quality. These columns are fundamental for any price-related analysis, and rows lacking this information would be unusable.
